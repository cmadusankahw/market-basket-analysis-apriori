```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
data=pd.read_excel("Online_etail.xlsx")
```

```python
data.head()
```

```python
data["Country"].value_counts()
```

```python
data_ger=data[data["Country"]== "Germany"]
```

```python
data_ger
```

```python
data_ger["InvoiceNo"]=data_ger["InvoiceNo"].astype(str)
```

```python
bkt = data_ger.groupby(["InvoiceNo","Description"])["Quantity"].sum().unstack()
```

```python
bkt = bkt.fillna(0)
```

```python
bkt
```

```python
def encode_units(x):
    if x<=0:
        return 0
    else:
        return 1
```

```python
# applymap -> apply a function to all values in a dataframe
bkt = bkt.applymap(encode_units)
```

```python
bkt
```

```python
# minimum support to include -> here 0.07
frq_items= apriori(bkt,min_support=0.07,use_colnames=True)
```

```python
frq_items
```

```python
rules=association_rules(frq_items,metric="lift",min_threshold=1)
```

```python
rules
```

```python
rules[(rules["lift"]>=2.5) & (rules["confidence"]>0.5)]
```
